primer Código de Agentes. Este código define a los dos primeros especialistas siguiendo los roles finanzas y legal:

nota: Una cosa importante antes de pegar el código: esa API key es visible públicamente ahora mismo en esta conversación. Te recomiendo que cuando termines las prácticas la regeneres en la web de Groq (console.groq.com → API Keys → crear nueva). Por ahora la usamos para el experimento.

```python
import os
from crewai import Agent, Task, Crew

# Configuración de la "Mente" de los agentes
os.environ["GROQ_API_KEY"] = "MI_GROQ_KEY"
os.environ["OTEL_SDK_DISABLED"] = "true"

# 1. Definición del Agente Analista Financiero
# Este agente se encarga de la división del trabajo y especialización [cite: 14]
analista_finanzas = Agent(
  role='Analista Financiero de Compras',
  goal='Maximizar el ahorro y asegurar que toda compra esté dentro del presupuesto empresarial',
  backstory='Eres un experto en finanzas con 10 años de experiencia. Tu prioridad es el control de costes.',
  verbose=True,
  allow_delegation=False,
  llm='groq/llama-3.3-70b-versatile' # <--- MODELO ACTUALIZADO
)

# 2. Definición del Agente de Compliance Legal
# Este agente añade una capa de razonamiento para mitigar riesgos [cite: 28]
especialista_legal = Agent(
  role='Auditor de Cumplimiento Legal',
  goal='Mitigar riesgos legales y verificar la fiabilidad de los proveedores',
  backstory='Tu objetivo es asegurar que la empresa cumpla con la normativa vigente.',
  verbose=True,
  allow_delegation=False,
  llm='groq/llama-3.3-70b-versatile' # <--- MODELO ACTUALIZADO
)

print("¡Agentes creados con éxito! Tu departamento virtual de Finanzas y Legal está listo.")
```

Significa que el "departamento virtual" ya tiene cerebro y está listo para recibir órdenes.

Fase de Ejecución de Tareas. el proceso de toma de decisiones

```python
# 1. Definición de la Misión (La Tarea)
# Aquí es donde el sistema demuestra su capacidad de análisis crítico
tarea_analisis_compra = Task(
    description="""Analizar la solicitud de compra de 50 licencias de un nuevo software de gestión logística. 
    El costo es de 10.000€ (un 20% más barato que el actual), pero el proveedor es una startup que solo lleva 6 meses operando.""",
    expected_output="""Un informe breve con una recomendación final (Comprar/No comprar). 
    Debe incluir un punto de vista financiero (ahorro) y uno legal (riesgo de continuidad).""",
    agent=analista_finanzas # El analista financiero inicia el proceso
)

# 2. Formación del Equipo (El Crew)
# Aquí es donde ocurre la magia de la colaboración que pide tu publicación Q1
departamento_compras = Crew(
    agents=[analista_finanzas, especialista_legal],
    tasks=[tarea_analisis_compra],
    verbose=True # IMPORTANTE: Esto te permite ver el "chat" entre los agentes en pantalla
)

# 3. ¡Lanzamiento de la simulación!
print("Lanzando simulación... observa cómo debaten los agentes:")
resultado = departamento_compras.kickoff()

# 4. Resultado final para el reporte
print("\n\n########################")
print("## RECOMENDACIÓN FINAL ##")
print("########################\n")
print(resultado)
```

Analizando tu éxito para el reporte:
- Entrada del Agente (Agent Started): El sistema ha aceptado la misión de analizar la compra de las 50 licencias.

- Pensamiento Crítico: Si te fijas en el texto, el agente está evaluando los pros (ahorro del 20%) y los contras (startup de 6 meses).

- Uso de Llama 3.3: El error anterior desapareció, lo que confirma que la migración al modelo llama-3.3-70b-versatile fue exitosa.
