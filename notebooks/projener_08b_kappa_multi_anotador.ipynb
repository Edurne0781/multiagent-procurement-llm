## 2. Cálculo de Cohen's Kappa por pares

```python
from sklearn.metrics import cohen_kappa_score
from itertools import combinations
import statistics

nombres = list(anotadores.keys())

print('=' * 65)
print('  COHEN\'S KAPPA POR PARES DE ANOTADORES')
print('=' * 65)
print(f'  Interpretación: <0.2 mínimo | 0.4-0.6 moderado | >0.8 casi perfecto')
print('-' * 65)

kappas = []
for a1, a2 in combinations(nombres, 2):
    y1 = [anotadores[a1][s] for s in signals]
    y2 = [anotadores[a2][s] for s in signals]
    k = cohen_kappa_score(y1, y2)
    kappas.append(k)
    nivel = '✅ casi perfecto' if k >= 0.8 else ('✅ sustancial' if k >= 0.6 else ('⚠️  moderado' if k >= 0.4 else '❌ bajo'))
    print(f'  {a1:25s} vs {a2:25s} κ = {k:.3f}  {nivel}')

print('-' * 65)
kappa_medio = statistics.mean(kappas)
kappa_min   = min(kappas)
kappa_max   = max(kappas)
print(f'  Kappa medio global:  κ = {kappa_medio:.3f}')
print(f'  Rango:               [{kappa_min:.3f} — {kappa_max:.3f}]')
print('=' * 65)
```

## 3. Matriz de acuerdo por señal

Muestra para cada señal cuántos anotadores coinciden y cuál es la clasificación mayoritaria.

```python
from collections import Counter

descripciones = {
    'S01': 'conflicto_interes',        'S02': 'clausula_exclusividad',
    'S03': 'gdpr_violation',           'S04': 'aesa_required',
    'S05': 'ce_certification',         'S06': 'importe_consejo',
    'S07': 'consignment_sin_contrato', 'S08': 'datos_en_riesgo',
    'S09': 'sistema_critico',          'S10': 'incidente_seguridad',
    'S11': 'ransomware',               'S12': 'pep_flag',
    'S13': 'aml_alert',                'S14': 'ofac_sanction',
    'S15': 'ubo_sancionado',           'S16': 'reach_violation',
    'S17': 'fraude_facturacion',       'S18': 'accidente_laboral',
    'S19': 'ofac_alert',               'S20': 'adquisicion_empresa',
    'S21': 'expansion_internacional',  'S22': 'desinversion_laboral',
    'S23': 'endeudamiento_significativo'
}

print('=' * 75)
print('  ACUERDO POR SEÑAL (n=6 anotadores)')
print('=' * 75)
print(f'  {"Señal":5s} {"Descripción":30s} {"Mayoritaria":12s} {"Acuerdo":8s} {"Detalle"}')
print('-' * 75)

senales_conflicto = []
for s in signals:
    votos = [anotadores[a][s] for a in nombres]
    contador = Counter(votos)
    mayoritaria, n_votos = contador.most_common(1)[0]
    pct = n_votos / len(nombres) * 100
    detalle = ' '.join([f"{v}({c})" for v, c in contador.most_common()])
    icono = '✅' if n_votos >= 5 else ('⚠️ ' if n_votos >= 4 else '❌')
    print(f'  {s:5s} {descripciones[s]:30s} {mayoritaria:12s} {n_votos}/6={pct:4.0f}% {icono} {detalle}')
    if n_votos < 5:
        senales_conflicto.append(s)

print('=' * 75)
print(f'  Señales con acuerdo unánime (6/6): {sum(1 for s in signals if Counter([anotadores[a][s] for a in nombres]).most_common(1)[0][1] == 6)}')
print(f'  Señales con posible conflicto (<5/6): {len(senales_conflicto)} → {senales_conflicto}')
```
